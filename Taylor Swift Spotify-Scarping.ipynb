```python
import requests
from bs4 import BeautifulSoup
import openpyxl

scrapbooks = openpyxl.Workbook()
sheet = scrapbooks.active
sheet.title = "DSP"

sheet.append(['Title'])

url = "https://artists.spotify.com/songwriter/1AHTiowIWg25JvwX4wr1MR"

try:
  req = requests.get(url)
  req.raise_for_status()

  soup = BeautifulSoup(req.text,'html.parser')

  books=soup.find_all('a') 

  for book in books:
    
    price = book.find('a',class_="Link-k8gsk-0 efSIwO styled__TitleLink-sc-t9ci4y-7 styled__RecordingTitle-sc-t9ci4y-8 jOOizS colGZg").a.text

    print(price)
    sheet.append([price])

except Exception as e:
    print(e)

scrapbooks.save('books.xlsx')
```

```python
import requests
from bs4 import BeautifulSoup
  
URL = "https://artists.spotify.com/songwriter/1AHTiowIWg25JvwX4wr1MR"
r = requests.get(URL)
  
soup = BeautifulSoup(r.content, 'html5lib') # If this line causes an error, run 'pip install html5lib' or install html5lib
print(soup.prettify())
```

```python
soup.find_all('a')
```

```python
names=soup.find_all('div',class_='styled__RecordingInfo-sc-t9ci4y-6 hBywMQ')
names
```

```python
streams=soup.find_all('div',class_="styled__ListCell-sc-t9ci4y-4 RecordingList__StreamCountCell-sc-ucdcer-2 eeZNwF fgnEym")
streams
```

```python
for name in names:
    k=name.find('a').text
for stream in streams:
    p=stream.find('span').text
print([k,p])
```

```python
p=soup.find('div',class_='RecordingList__TableContainer-sc-ucdcer-1 fgofhZ')
p
```

```python
SongStat=p.find_all('div',class_='styled__TableGrid-sc-t9ci4y-1 ifLTnm')
SongStat
```

```python
for ss in SongStat:
    if(ss.find('a',class_='Link-k8gsk-0 efSIwO styled__TitleLink-sc-t9ci4y-7 styled__RecordingTitle-sc-t9ci4y-8 jOOizS colGZg')):
        k=ss.find('a',class_='Link-k8gsk-0 efSIwO styled__TitleLink-sc-t9ci4y-7 styled__RecordingTitle-sc-t9ci4y-8 jOOizS colGZg').text
        p=ss.find('div',class_="styled__ListCell-sc-t9ci4y-4 RecordingList__StreamCountCell-sc-ucdcer-2 eeZNwF fgnEym").text
        print([k,p])
```

```python
names=p.find_all('a',class_='Link-k8gsk-0 efSIwO styled__TitleLink-sc-t9ci4y-7 styled__RecordingTitle-sc-t9ci4y-8 jOOizS colGZg')  
names
```

```python
streams=p.find_all('div',class_="styled__ListCell-sc-t9ci4y-4 RecordingList__StreamCountCell-sc-ucdcer-2 eeZNwF fgnEym")
streams
```

```python
Song=[]
Stat=[]

for name in names:
    Song.append(name.text)
for stream in streams:
    Stat.append(stream.text)
print([Song,Stat])
```
